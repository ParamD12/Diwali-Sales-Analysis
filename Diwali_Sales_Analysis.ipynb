```python
# Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Load the Dataset

df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
# Drop Irrelevant Columns
 
df.drop(['Status', 'unnamed1'], axis = 1, inplace = True)
```

```python
# Drop Null Values

df.dropna(inplace = True)
```

```python
# Sales and Revenue Contribution By Gender
```

```python
# Count of Gender 
 
ax = sns.countplot(x= 'Gender', data= df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Revenue Contribution Per Gender

sales_gen = df.groupby('Gender', as_index=False)['Amount'].sum().sort_values('Amount', ascending = True)

ax = sns.barplot(x= 'Gender', y='Amount', hue= 'Gender', data= sales_gen)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Sales and Revenue Contribution By Age Group
```

```python
ax = sns.countplot(x = 'Age Group', hue = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(by = 'Age Group', ascending=True)

sns.set(rc={'figure.figsize':(10, 5)})

ax = sns.barplot(x = 'Age Group', y='Amount', data = sales_age)

# for bars in ax.containers:
#     ax.bar_label(bars)
```

```python
# Using a Pie-Chart

sales_age2 = df.groupby('Age Group')['Amount'].sum()

# Calculate percentage contribution to revenue
sales_age_percentage = (sales_age2 / sales_age2.sum()) * 100

plt.figure(figsize=(10,10))

plt.pie(sales_age2, autopct=lambda p: '{:.1f}%'.format(p))

plt.legend(sales_age2.index, title = 'Age Group')

plt.title('Reveue Contribution Per Age Group')

plt.show()
```
